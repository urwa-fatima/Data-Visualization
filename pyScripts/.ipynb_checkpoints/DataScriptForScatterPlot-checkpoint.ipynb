```python
# All the import Libraries
import pandas as pd
import geopandas as gp
from matplotlib.patches import Polygon
import matplotlib.pyplot as plt
```

```python
# read files
data = gp.read_file("../data/geo_data_trees.geojson")
polydata = gp.read_file("../data/circoscrizioni.json")
```

```python
# deleting the last row
data = data.drop(data.index[-1])
```

```python
# the last row is the total row for all the data so should be avoided 
# when including in the data values.
data.tail()
```

```python
# 12513 rows × 26 columns
# Basic information about all the columns of the data
data.info()
```

```python
data["count"]=1
```

```python
top_tree_type = data[['Name', 'count']][:-1]
top_tree_type
```

```python
top_tree = top_tree_type.groupby("Name").count().reset_index()
# top_tree_type.sort_values(by="count", ascending=False)[:6]
top_tree = top_tree.sort_values(by='count', ascending=False) 
Top_six_Trees = list(top_tree[:6].Name)
Top_six_Trees
```

```python
# For Getting the values of Top 6 Tree types
df = data[["Name", "DBH (cm)","Height (m)","Crown Height (m)","Crown Width (m)", "Canopy Cover (m2)","Gross Carbon Sequestration (kg/yr)", "Gross Carbon Sequestration (eur/yr)" ]]
```

```python
df
```

```python
# return values filtered based on Top 6 tree types fron the dataframe
data_for_top_six = df[df["Name"].isin(Top_six_Trees)]
```
