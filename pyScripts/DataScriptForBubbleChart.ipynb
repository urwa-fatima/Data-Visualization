```python
bubble_data = data[['Name', 'DBH (cm)','Carbon Storage (kg)', 'Canopy Cover (m2)']]
bubble_data
```

```python
bubble_data.columns = ['Name', 'Size', 'Carbon', 'Canopy']
bubble_data[['Size', 'Carbon', 'Canopy']] = bubble_data[['Size', 'Carbon', 'Canopy']].apply(pd.to_numeric)
```

```python
bubble_data.value_counts('Name')
```

```python
grouped_data = bubble_data[[ 'Size', 'Carbon', 'Canopy','Name']].groupby('Name').mean()
grouped_data
```

```python
grouped_data.to_csv (r'/content/bubble_chart_data.csv', index = True, header=True)
```
