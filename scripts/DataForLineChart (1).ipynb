```python
import pandas as pd
```

```python
mydata = pd.read_excel("../data/data.xlsx")
mydata
```

```python
# iterating the columns
for col in mydata.columns:
    print(col)
# Delect rows based on multiple column value
mydata = mydata[(mydata['min'] != '---') & (mydata['max'] != '---')]
mydata=mydata.drop([14975])
mydata
```

```python
mydata.index = pd.to_datetime(mydata['dates'],format='%d.%m.%Y')
mydata['day'] = mydata.index.day
mydata['month'] = mydata.index.month
mydata['year'] = mydata.index.year
mydata.head()
```

```python
mydata=mydata.drop(['dates'], axis=1)
mydata.reset_index()
```

```python
mydata.dtypes
```

```python
mydata['max'] = mydata['max'].astype(float)
mydata['min'] = mydata['min'].astype(float)
```

```python
df = mydata.groupby(['year','month','dates']).agg(min_temp=('min', 'mean'), 
                               max_temp=('max', 'mean'))
df
```

```python
df = df.reset_index()
```

```python
df['avg'] = (df['min_temp']+df['max_temp'])/2
df
```

```python
import calendar
df['Month'] = df['month'].apply(lambda x: calendar.month_abbr[x])
```
